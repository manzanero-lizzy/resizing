```python
#!/usr/bin/python
from PIL import Image
import os, sys

#directory containing fish images - palitan nyo nalang
#note: dapat forward slash
path = "C:/Users/Eliza Marie/Documents/4th year first sem/not a fish/"
dirs = os.listdir( path )

#the output will be placed to "yourpathtofishimages/resized/"

def resize():
	i = 1
	for item in dirs:
		if os.path.isfile(path+item):

			if not item.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
				i+=1
				continue
			
			im = Image.open(path+item)
			f, e = os.path.splitext(path+item)
			filename = item.split(".")[0]
			# exist or not.
			if not os.path.exists(path+"resized/"):
				# if the demo_folder directory is not present 
				# then create it.
				os.makedirs(path+"resized/")

			print(f"    {i} - resizing {item}")
			imResize = im.resize((256,256), Image.ANTIALIAS)
			imResize.save(path + 'resized/' + filename + ' 256 x 256.jpg', 'JPEG', quality=100, subsampling=0)
		i+=1

resize()
print("\n    done")
```
